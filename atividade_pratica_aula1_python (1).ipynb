## 7) Atividade 5 — Escolhendo a visualização certa

Sem necessariamente construir os gráficos ainda, responda:

Em seguida, gere **ao menos um gráfico** com `pandas` ou `matplotlib`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("vendas_brasil_aula1.csv")

receita_canal = df.groupby("Canal_Venda")["Receita"].sum()

plt.figure(figsize=(8,5))
receita_canal.plot(kind='bar', color='skyblue')

plt.title("Receita por Canal de Venda")
plt.xlabel("Canal de Venda")
plt.ylabel("Receita Total")
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

## Desafio extra (opcional)

Faça uma pequena limpeza no dataset:

- trate valores nulos;
- padronize os nomes dos canais de venda;
- converta `Data_Venda` para formato de data.

Depois, compare os resultados **antes e depois da limpeza**.

```python
import pandas as pd

df = pd.read_csv("vendas_brasil_aula1.csv")

df["Canal_Venda"] = df["Canal_Venda"].fillna("Desconhecido")

df["UF"] = df["UF"].fillna("Desconhecido")

df["Preco_Unitario"] = df["Preco_Unitario"].fillna(df["Preco_Unitario"].mean())

df["Canal_Venda"] = df["Canal_Venda"].str.lower().str.strip()

df["Canal_Venda"] = df["Canal_Venda"].replace({
    "e-commerce": "e-commerce",
    "loja fisica": "loja física",
    "loja física": "loja física"
})

df["Canal_Venda"] = df["Canal_Venda"].replace({
    "e-commerce": "E-commerce",
    "loja física": "Loja Física",
    "marketplace": "Marketplace",
    "whatsapp": "WhatsApp",
    "desconhecido": "Desconhecido"
})

df["Data_Venda"] = pd.to_datetime(df["Data_Venda"])

print("Valores nulos após limpeza:")
print(df.isnull().sum())

print("\nCanais padronizados:")
print(df["Canal_Venda"].value_counts())
```

**Por que uma análise de dados não deve se limitar a “fazer gráficos bonitos”?**

Uma análise de dados não deve se limitar a “fazer gráficos bonitos” porque gráficos apenas ilustram informações — eles não garantem que os dados estejam corretos ou que as conclusões sejam válidas. Sem tratamento e validação, é possível gerar visualizações atraentes baseadas em dados inconsistentes, levando a decisões estratégicas equivocadas. O papel do analista é entender o contexto, garantir a qualidade dos dados e interpretar os resultados com senso crítico. Assim, o valor da análise está na capacidade de gerar insights confiáveis que apoiem decisões de negócio, e não apenas na estética das visualizações.
